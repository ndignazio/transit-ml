```python
features = small_df.drop(columns=['GEO_ID', 'tract', 'commuting_ridership'], axis=1)
```

```python
num_to_add = 0

threshold = 100 - num_to_add


features_plus_0 = features.copy(deep=True)
features_plus_0['transit_score'] = features_plus_0['transit_score'] + num_to_add
features_plus_0['transit_score'][features_plus_0['transit_score'] > 100] = 100

num_to_add = 5
features_plus_5 = features.copy(deep=True)
features_plus_5['transit_score'] = features_plus_5['transit_score'] + num_to_add
features_plus_5['transit_score'][features_plus_5['transit_score'] > 100] = 100

num_to_add = 10
features_plus_10 = features.copy(deep=True)
features_plus_10['transit_score'] = features_plus_10['transit_score'] + num_to_add
features_plus_10['transit_score'][features_plus_10['transit_score'] > 100] = 100

num_to_add = 15
features_plus_15 = features.copy(deep=True)
features_plus_15['transit_score'] = features_plus_15['transit_score'] + num_to_add
features_plus_15['transit_score'][features_plus_15['transit_score'] > 100] = 100
    
```

```python
poly_features_plus_0 = pd.DataFrame(poly.fit_transform(features_plus_0))
poly_features_plus_5 = pd.DataFrame(poly.fit_transform(features_plus_5))
poly_features_plus_10 = pd.DataFrame(poly.fit_transform(features_plus_10))
poly_features_plus_15 = pd.DataFrame(poly.fit_transform(features_plus_15))
```

```python
plus_0_predvals = best_model.predict(poly_features_plus_0)
plus_5_predvals = best_model.predict(poly_features_plus_5)
plus_10_predvals = best_model.predict(poly_features_plus_10)
plus_15_predvals = best_model.predict(poly_features_plus_15)
```

```python
small_df['commuting_ridership_pred_tscore_plus0'] = plus_0_predvals
small_df['commuting_ridership_pred_tscore_plus5'] = plus_5_predvals
small_df['commuting_ridership_pred_tscore_plus10'] = plus_10_predvals
small_df['commuting_ridership_pred_tscore_plus15'] = plus_15_predvals
small_df['plus0_diff'] = small_df['commuting_ridership'] - \
                            small_df['commuting_ridership_pred_tscore_plus0'] 

small_df['plus5_diff'] = small_df['commuting_ridership_pred_tscore_plus5'] - \
                            small_df['commuting_ridership_pred_tscore_plus0']
small_df['plus10_diff'] = small_df['commuting_ridership_pred_tscore_plus10'] - \
                            small_df['commuting_ridership_pred_tscore_plus0']
small_df['plus15_diff'] = small_df['commuting_ridership_pred_tscore_plus15'] - \
                            small_df['commuting_ridership_pred_tscore_plus0']
```

```python
results_df = small_df[['GEO_ID', 'transit_score', 'commuting_ridership', 'commuting_ridership_pred_tscore_plus0',
                       'plus0_diff', 'commuting_ridership_pred_tscore_plus5', 'plus5_diff',
                       'commuting_ridership_pred_tscore_plus10', 'plus10_diff',
                      'commuting_ridership_pred_tscore_plus15', 'plus15_diff']]
```
